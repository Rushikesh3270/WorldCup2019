# Indian Team Data

```python
Indian_players=players[players['Country']=="India"]
```

```python
Indian_players
```

```python
India = odi_total[odi_total['Country']=='India']
```

```python
India_win = India[India['Result']=='won']
```

```python
India_win['Opposition'].value_counts()
```

```python
plt.figure(figsize=(10,6))
India_win['Opposition'].value_counts().plot(kind='bar')
plt.title('Most wins Against Respective Teams',fontsize=20)
plt.xlabel('Team Names')
plt.ylabel('Wins');
```

```python
India_loss = India[India['Result']=='lost']
```

```python
India_loss['Opposition'].value_counts()
```

```python

plt.figure(figsize=(9,9))
India_loss['Opposition'].value_counts().plot(kind='pie',autopct='%0.1f%%');
```

```python
India[India.Target>300]   
```

```python
India[India.Target>300].Result.value_counts().plot(kind='bar')
plt.title('Matches Won While Chasing 300+ Score');
```

```python
India.Ground.value_counts()
```

```python
plt.figure(figsize=(17,7))
India.Ground.value_counts().plot(kind='bar')
plt.title('Most Matches Played On Each Venue',fontsize=20)
plt.xlabel('Venue',fontsize=15);
```

```python
India[India.Target>400].Result.value_counts().plot(kind='bar');
```

```python
plt.figure(figsize=(9,6))
India[India.Result=='won'].Ground.value_counts().plot(kind='bar')
plt.title('Most Wins On Each Venus')
plt.xlabel('Venus');
```
